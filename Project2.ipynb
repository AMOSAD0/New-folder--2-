```python
class book:
  def __init__(self,title,author,numCopies,):
    self.title=title
    self.author=author
    self.numCopies=numCopies
    self.numBorrow=0
    self.numAvailable=numCopies

  def borrowBook(self):
    self.numBorrow+=1
    self.numAvailable-=1
    print('Number of copies available',self.numAvailable)
  def returnBook(self):
    self.numBorrow-=1
    self.numAvailable+=1
    print('Number of copies available',self.numAvailable)
    print('Number of copies borrowed : ',self.numBorrow)
  def infoBook(self):
    print('title : '+self.title)
    print('-'*25)
    print('author : '+self.author)
    print('-'*25)
    print('Number of copies available :', self.numAvailable)
    print('-'*25)
    print('Number of copies borrowed : ',self.numBorrow)
    print('-'*25)
    print('Number of copies : ',self.numCopies)
    print('-'*25)

class library:
  def __init__(self):
    self.listLibrary={}
  def addBook(self,book):
    self.listLibrary.update({book.title:book})
  def removeBook(self,book):
    self.listLibrary.pop(book.title)
  def searchBook(self,t):
    try:
      self.listLibrary[t].infoBook()
      return  self.listLibrary[t]
    except:
      print('not found')
  def printLibrary(self):
    print(self.listLibrary.keys())


book1= book(title='A',author='ahmed',numCopies=10)
#book1.infoBook()
book2= book(title='B',author='ahmed',numCopies=20)
#book2.infoBook()
book3= book(title='C',author='ahmed',numCopies=30)
book3.infoBook()

lib= library()

lib.addBook(book1)
lib.addBook(book2)
lib.addBook(book3)

lib.printLibrary()

lib.removeBook(book1)

lib.printLibrary()

bk=lib.searchBook(book1.title)

bk=lib.searchBook(book2.title)

bk.borrowBook()

bk.returnBook()
```
